## Гистограмма (частотная диаграмма, столбчатая диаграмма)

Разновидность научной и инжинерной графики, которая позволяет визуально оценить базовые параметры распределения случайных величин.
По оси абсцисс откладывается значение параметра, по оси ординат число проб (замеров), находящихся в диапазоне значений. 
Гистограмма позволяет оценить параметры распределения (мин, среднее, макс, среднеквадратическое отклонение), его симметрию, баланс и т.д.


### Гистограмма как индикатор экспозиции изображения

Построить гистограмму изображения библиотеки scikit-image



```python
from skimage import data
import numpy as np

img = data.camera()
img = np.asarray(img,dtype=np.uint8)

def img_hist(img):
    img_flat = img.flatten() #разворачивает прямоугольную матрицу в вектор
    fig, (ax1, ax2) = plt.subplots(1, 2)
    plt.subplot(1,2,1)
    ax1.imshow(img,cmap='gray',aspect='auto')
    plt.subplot(1,2,2)
    ax2.hist(img_flat,color="skyblue",     # Fill color
                 edgecolor="black",   # Border color
                 linewidth=0.3)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

    plt.show()

img_hist(img)

print('Усилим яркость изображения')
k=10

img_2x = np.multiply(img, k, dtype=np.uint8)

img_hist(img_2x)

img_05x = img
img_05x[:200,:100] = img_05x[:200,:100] - img_05x[:200,:100]//2
img_hist(img_05x)


```
